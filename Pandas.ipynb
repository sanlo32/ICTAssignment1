1. Basic Exploration
Load the dataset into a Pandas DataFrame. Display:
○ The first 10 rows
○ The total number of rows and columns
○ Summary statistics for numerical columns

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\asnsu\ICT\Assignment 1\auto-mpg.csv")
display(df.head(10))
print("Shape (rows, columns):", df.shape)
display(df.describe())
```

Filtering and Indexing
Find all cars manufactured in 1975 with a weight less than 3000. Return the
DataFrame with selected columns: car_name, weight, and mpg.


```python
filtered_df = df[(df["model year"] == 75) & (df["weight"] < 3000)][
    ["car name", "weight", "mpg"]
]
display(filtered_df)
```

Handling Missing Data
Identify if there are any missing values in the dataset. Replace missing values in the
horsepower column with the column's median.


```python
missing_values = df.isnull().sum()
print("Missing values per column:\n", missing_values)
df["horsepower"] = pd.to_numeric(df["horsepower"], errors="coerce")
median_hp = df["horsepower"].median()
df["horsepower"] = df["horsepower"].fillna(median_hp)
print("Median horsepower:", median_hp)
```

4. Data Transformation
Add a new column power_to_weight_ratio, calculated as horsepower / weight.

```python
df["power_to_weight_ratio"] = df["horsepower"] / df["weight"]
display(df[["horsepower", "weight", "power_to_weight_ratio"]].head())
```

5. Group By
Group the cars by origin and calculate the mean mpg for each group

```python
mean_mpg_by_origin = df.groupby("origin")["mpg"].mean()
display(mean_mpg_by_origin)
```

Sorting
Sort the DataFrame by mpg in descending order and display the top 10 cars with
the highest mpg.
Apply Function
Create a new column performance_score using a custom function:
def performance_score(row):
return row['mpg'] * row['acceleration'] / row['weight']


```python
def performance_score(row):
    return row["mpg"] * row["acceleration"] / row["weight"]


top10_mpg = df.sort_values(by="mpg", ascending=False).head(10)
display(top10_mpg[["car name", "mpg", "acceleration", "weight"]])
```

Apply this function to each row and store the result in the new column.

```python
df["performance_score"] = df.apply(performance_score, axis=1)
display(df[["car name", "mpg", "acceleration", "weight", "performance_score"]].head())
```

Visualization Preparation
Generate a summary DataFrame with:
○ Average mpg, weight, and horsepower for each model_year

```python
summary_by_year = (
    df.groupby("model year")[["mpg", "weight", "horsepower"]].mean().reset_index()
)
summary_by_year.rename(
    columns={"mpg": "avg_mpg", "weight": "avg_weight", "horsepower": "avg_horsepower"},
    inplace=True,
)
display(summary_by_year)
```

Exporting Data
Save a subset of the data containing only mpg, cylinders, horsepower, and weight
for cars with mpg > 30 into a CSV file named high_mpg_cars.csv.

```python
high_mpg_cars = df[df["mpg"] > 30][["mpg", "cylinders", "horsepower", "weight"]]
high_mpg_cars.to_csv("high_mpg_cars.csv", index=False)
```
