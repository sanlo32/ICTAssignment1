1. Basic Array Operations
Convert the mpg column into a NumPy array and calculate:
○ The mean, median, and standard deviation of mpg.
○ The number of cars with mpg greater than 25.

```python
mpg_array = df["mpg"].values
```

```python
mpg_array
```

```python
import numpy as np

mpg_mean = np.mean(mpg_array)
mpg_median = np.median(mpg_array)
mpg_std = np.std(mpg_array)

print("Mean:", mpg_mean)
print("Median:", mpg_median)
print("Standard Deviation:", mpg_std)
```

```python
num_cars_gt_25 = np.sum(mpg_array > 25)
print("Number of cars with mpg greater than 25:", num_cars_gt_25)
```

Filtering
Using NumPy, filter all cars with more than 6 cylinders.
Return the corresponding car_name as a list

```python
car_name = df["car name"].values[df["cylinders"].values > 6].tolist()
car_name
```

Statistical Analysis
Compute the 25th, 50th, and 75th percentiles of the weight column using NumPy

```python
weight_array = df["weight"].values
q1 = np.percentile(weight_array, 25)
q2 = np.percentile(weight_array, 50)
q3 = np.percentile(weight_array, 75)

print("25th percentile:", q1)
print("50th percentile:", q2)
print("75th percentile:", q3)
```

Array Manipulation
Convert the acceleration column into a NumPy array and normalize its values
(scale between 0 and 1).

```python
acceleration_array = df["acceleration"].values
acceleration_norm = (acceleration_array - acceleration_array.min()) / (
    acceleration_array.max() - acceleration_array.min()
)
acceleration_norm
```

Broadcasting
Increase all horsepower values by 10% and store the updated values in a new
NumPy array. Handle missing data (if any) by replacing it with the mean of the
column before applying the increase.

```python
horsepower = pd.to_numeric(df["horsepower"], errors="coerce")

mean_hp = np.nanmean(horsepower)
horsepower_filled = np.where(np.isnan(horsepower), mean_hp, horsepower)

horsepower_updated = horsepower_filled + horsepower_filled * 0.1
horsepower_updated
```

Boolean Indexing
Find the average displacement of cars with an origin of 2 (Europe) using NumPy
indexing

```python
avg_displacement_europe = np.mean(df["displacement"].values[df["origin"].values == 2])
print("Average displacement of cars with origin 2 (Europe):", avg_displacement_europe)
```

Matrix Operations
Create a 2D NumPy array containing the columns mpg, horsepower, and weight.
Compute the dot product of this matrix with a given vector [1, 0.5, -0.2].


```python
matrix = np.column_stack((mpg_array, horsepower_filled, weight_array))
vector = np.array([1, 0.5, -0.2])
dot_product = np.dot(matrix, vector)
dot_product
```

Use NumPy to sort the cars by model_year in descending order and display the first

Correlation
Compute the Pearson correlation coefficient between mpg and weight using
NumPy.

```python
corr_coef = np.corrcoef(mpg_array, weight_array)[0, 1]
print("Pearson correlation coefficient between mpg and weight:", corr_coef)
```

Calculate the mean mpg for cars grouped by the number of cylinders using NumPy
techniques.

```python
unique_cylinders = np.unique(df["cylinders"].values)
mean_mpg_by_cyl = {
    cyl: np.mean(mpg_array[df["cylinders"].values == cyl]) for cyl in unique_cylinders
}
mean_mpg_by_cyl
```
